```python
import pandas as pd
df = pd.read_csv ("FAOSTAT_DATA_7-23-2022.CSV")
print(df.columns)
df.head()
                  
```

```python
# Filter only Ghana and Côte d'Ivoire
df_filtered = df[df["Area"].isin(["Ghana", "Côte d'Ivoire"])]

# Pivot so Elements become columns
df_pivot = df_filtered.pivot_table(
    index=["Area", "Year"],
    columns="Element",
    values="Value"
).reset_index()

# Rename columns for simplicity
df_pivot = df_pivot.rename(columns={
    "Area": "Country",
    "Area harvested": "Area harvested",
    "Yield": "Yield",
    "Production": "Production"
})

# Split into Ghana and Côte d'Ivoire tables
ghana = df_pivot[df_pivot["Country"] == "Ghana"][["Year", "Area harvested", "Yield", "Production"]]
ivory = df_pivot[df_pivot["Country"] == "Côte d'Ivoire"][["Year", "Area harvested", "Yield", "Production"]]

print("Ghana Table:")
print(ghana.head())

print("\nCôte d'Ivoire Table:")
print(ivory.head())
```

```python
import matplotlib.pyplot as plt

# Ghana scatter
plt.scatter(ghana["Year"], ghana["Yield"], color="green")
plt.title("Ghana - Yield by Year")
plt.xlabel("Year")
plt.ylabel("Yield (hg/ha)")
plt.show()

# Côte d'Ivoire scatter
plt.scatter(ivory["Year"], ivory["Yield"], color="brown")
plt.title("Côte d’Ivoire - Yield by Year")
plt.xlabel("Year")
plt.ylabel("Yield (hg/ha)")
plt.show()
```

```python
plt.bar(ghana["Year"] ,ghana["Area harvested"],color="green")
plt.title("Ghana-Area harvested by year")
plt.xlabel("Year")
plt.ylabel("Area harvested (ha)")
plt.show()

plt.bar(ivory["Year"] ,ivory["Area harvested"],color="blue")
plt.title("Ivory-Area harvested by year")
plt.xlabel("Year")
plt.ylabel("Area harvested (ha)")
plt.show()
```

```python
fig , axes = plt.subplots(2,2 , figsize=(12,8))
# Ghana scatter
axes[0,0].scatter(ghana["Year"], ghana["Yield"], color="green")
axes[0,0].set_title("Ghana - Yield by Year")
axes[0,0].set_xlabel("Year")
axes[0,0].set_ylabel("Yield (hg/ha)")


# Côte d'Ivoire scatter
axes[0,1].scatter(ivory["Year"], ivory["Yield"], color="brown")
axes[0,1].set_title("Côte d’Ivoire - Yield by Year")
axes[0,1].set_xlabel("Year")
axes[0,1].set_ylabel("Yield (hg/ha)")

#Ghana bar
axes[1,0].bar(ghana["Year"] ,ghana["Area harvested"],color="green")
axes[1,0].set_title("Ghana-Area harvested by year")
axes[1,0].set_xlabel("Year")
axes[1,0].set_ylabel("Area harvested (ha)")

#ivory bar
axes[1,1].bar(ivory["Year"] ,ivory["Area harvested"],color="blue")
axes[1,1].set_title("Ivory-Area harvested by year")
axes[1,1].set_xlabel("Year")
axes[1,1].set_ylabel("Area harvested (ha)")

#title
fig.suptitle("Cocoa Production: Ghana vs Côte d'Ivoire", fontsize=16)

plt.tight_layout()
plt.savefig("cocoa_production.pdf")
plt.show()



```
